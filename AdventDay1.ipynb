```python
import re
#items = open('adventday1.txt','r').read().split('\n')
items= open('challenge 01 input 1.txt','r').read().splitlines()
```

```python
items
```

```python
numbers = list()
for x in items:
    match = re.findall(r'\d', x)
    if match:
        first = int(match[0])
        last = int(match[-1])
        num = first*10 +last
        numbers.append(num)
#print(numbers)
answer=sum(numbers)
print(answer)
```

```python
replaced = [
    ('one','one1one'),
    ('two','two2two'),
    ('three','three3three'),
    ('four','four4four'),
    ('five','five5five'),
    ('six','six6six'),
    ('seven','seven7seven'),
    ('eight','eight8eight'),
    ('nine','nine9nine')
]
number=0
for x in items:
    for old, new in replaced:
        x = x.replace(old, new)
    #print(x)
    match = re.findall(r'\d', x)
    #print(match)
    if match:
        first = int(match[0])
        last = int(match[-1])
    else:
        print('no match')
    num = first*10 +last
    number += num
    #print(num)
#print(numbers)
#answer=sum(number)
print(number)
```
