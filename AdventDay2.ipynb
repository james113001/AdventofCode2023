```python
import os
import pandas as pd
import re
```

```python
df = pd.read_csv('AdventDay2.txt', sep= ':',header=None)
print(df)
```

```python
df = df.set_axis(['Game','Values'], axis=1)
print(df)
```

```python
df['Game'] =df['Game'].str.replace('[^\d\.]','', regex=True).astype(int)

df['Blue'] = df['Values'].str.findall(r'(\d+) blue')
df['Blue'] = df['Blue'].apply(lambda lst: list(map(int, lst)))
df['red'] = df['Values'].str.findall(r'(\d+) red')
df['red'] = df['red'].apply(lambda lst: list(map(int, lst)))
df['green'] = df['Values'].str.findall(r'(\d+) green')
df['green'] = df['green'].apply(lambda lst: list(map(int, lst)))

df
```

```python
df['blueFLAG'] = df.Blue.explode().gt(14).groupby(level=0, sort=False).any()
df['redFLAG'] = df.red.explode().gt(12).groupby(level=0, sort=False).any()
df['greenFLAG'] = df.green.explode().gt(13).groupby(level=0, sort=False).any()

df["FLAG"] = df[['blueFLAG', 'redFLAG', 'greenFLAG']].any(axis='columns')
df
```

```python
filtered = df.loc[df['FLAG'] == False]
filtered
```

```python
sum(filtered['Game'])
```

```python
df['bluemax'] = df['Blue'].apply(lambda x: max(x))
df['redmax'] = df['red'].apply(lambda x: max(x))
df['greenmax'] = df['green'].apply(lambda x: max(x))
df['power'] = df['bluemax']*df['redmax']*df['greenmax']
print(sum(df['power']))
```
